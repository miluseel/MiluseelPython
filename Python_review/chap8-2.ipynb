## 클래스의 추가적인 구문

* isinstance() :어떤 클래스의 인스턴스인지 확인할 때 사용하는 함수
* 클래스변수와 클래스 함수는 클래스 이름 뒤에 마침표찍고 바로 사용 가능한 클래스가 갖고 있는 변수와 함수
* 상속 : 어떤 클래스를 기반으로 그 속성과 기능을 물려받아 새로운 클래스를 만드는 것

```python
#isinstance()활용

#학생클래스 선언
class Student:
    def study(self):
        print("공부를 합니다.")

#선생님 클래스 선언
class Teacher:
    def teach(self):
        print("학생을 가르칩니다.")

#교실 내부의 객체 리스트 생성
classroom = [Student(), Student(), Teacher(), Student(), Student()]

#반복적용하여 적절한 함수 호출
for person in classroom:
    if isinstance(person, Student):
        person.study()
    elif isinstance(person, Teacher):
        person.teach()
```

```python
#__str__()함수

#클래스 선언
class Student:
    def __init__(self, name, korean, math, english, science):
        self.name = name
        self.korean = korean
        self.math = math
        self.english = english
        self.science = science

    def get_sum(self):
        return self.korean + self.math +\
            self.english + self.science

    def get_average(self):
        return self.get_sum() / 4
    
    def __str__(self):
        return "{}\t{}\t{}".format(
            self.name,
            self.get_sum(),
            self.get_average())
    
#학생리스트 선언
students = [ 
    Student("윤인성", 87, 98, 88, 95),
    Student("연하진", 92, 98, 96, 98),
    Student("구지연", 76, 96, 94, 90),
    Student("나선주", 98, 92, 96, 92),
    Student("윤아린", 95, 98, 98, 98),
    Student("윤영월", 64, 88, 92, 92)
]

#출력
print("이름","총점", "평균", sep="\t")
for student in students:
    print(str(student))
```

### 클래스 함수<br>
: 클래스 변수 처럼 클래스를 가진 함수<br>
: 일반적인 함수/ 클래스함수에 사용은 차이가 없으나 '클래스가 가진 기능'을 명시적으로 나타냄<br>

* 클래스 함수 생성<br>
: class 클래스 이름:<br>
&emsp;@classmethod<br>
&emsp;def 클래스함수(cls, 매개변수):<br>
&emsp;&emsp;pass

* 클래스 함수 호출<br>
: 클래스이름. 함수이름(매개변수)


* 데코레이터 
    - 클래스 데코레이터
    - 함수 데코레이터

```python
#클래스 함수
class Student:
    #클래스 변수
    count = 0 
    students = []

    #클래스 함수
    @classmethod
    def print(cls):
        print("-----학생목록-----")
        print("이름\t총점\t평균")
        for student in cls.students:
            print(str(student))
        print("----- ----- -----")

    #인스턴스 함수
    def __init__(self, name, korean, math, english, science):
        self.name = name
        self.korean = korean
        self.math = math
        self.english = english
        self.science = science
        Student.count += 1
        Student.students.append(self)

    def get_sum(self):
        return self.korean + self.math +\
            self.english + self.science
    def get_average(self):
        return self.get_sum() / 4
    
    def __str__(self):
        return "{}\t{}\t{}".format(\
            self.name,\
            self.get_sum(),\
            self.get_average())
        

Student("윤인성", 87, 98, 88, 95),
Student("연하진", 92, 98, 96, 98),
Student("구지연", 76, 96, 94, 90),
Student("나선주", 98, 92, 96, 92),
Student("윤아린", 95, 98, 98, 98),
Student("윤영월", 64, 88, 92, 92),
Student("김미화", 87, 98, 88, 95),
Student("김연화", 92, 98, 96, 98),
Student("박아현", 76, 96, 94, 90),
Student("서준서", 98, 92, 96, 92),

#현재 생성된 학생 출력
Student.print()
```

### 프라이빗 변수와 게터/세터

```python
#원의 둘레와 넓이를 구하는 객체 지향 프로그램

#모듈
import math

#클래스 선언 
class Circle:
    def __init__(self,radius):
        self.radius = radius
    def get_circumference(self):
        return 2 * math.pi * self.radius
    def get_area(self):
        return math.pi * ( self.radius ** 2)
    
# 원의 둘레와 넓이
circle = Circle(10)
print("원의 둘레:", circle.get_circumference())
print("원의 넓이:", circle.get_area())
```

-> 길이를 음수로 넣는 것을 막는 방법 : 프라이빗 변수

```python
#프라이빗 변수

import math

#클래스 선언
class Circle:
    def __init__(self, radius):
        self.__radius = radius
    def get_circumference(self):
        return 2 * math.pi * self.__radius
    def get_area(self):
        return math.pi * (self.__radius ** 2)
    
# 원의 둘레와 넓이
circle = Circle(10)
print("원의 둘레:", circle.get_circumference())
print("원의 넓이:", circle.get_area())
print()

#__radius 에 접근
print(circle.__radius)
```

## 게터 / 세터<br>
: 프라이빗 변수의 값을 추출하거나 변경할 목적으로 간접적으로 송성에 접근하도록 해주는 함수

```python
#게터와 세터


#클래스 선언
class Circle:
    def __init__(self, radius):
        self.__radius = radius
    def get_circumference(self):
        return 2 * math.pi * self.__radius
    def get_area(self):
        return math.pi * (self.__radius ** 2)
    
#게터, 세터 선언
    def get_radius(self):
        return self.__radius
    def set_radius(self, value):
        self.__radius = value

   
# 원의 둘레와 넓이
circle = Circle(10)
print("원의 둘레:", circle.get_circumference())
print("원의 넓이:", circle.get_area())
print()

# 간접적으로 __radius 에 접근
print(circle.get_radius())
print()

#원의 둘레와 넓이
circle.set_radius(2)
print("원의 둘레:", circle.get_circumference())
print("원의 넓이:", circle.get_area())
```

### 데코레이터를 사용한 게터와 세터<br>
변수이름과 같은 함수 정의 후 위에 데코레이터 추가<br>
@property <br>
@<게터함수이름>.setter

```python
#데코레이터를 사용해 게터와 세터 만들기
import math


#클래스 선언
class Circle:
    def __init__(self, radius):
        self.__radius = radius
    def get_circumference(self):
        return 2 * math.pi * self.__radius
    def get_area(self):
        return math.pi * (self.__radius ** 2)
    

#게터, 세터 선언
    @property
    def radius(self):
        return self.__radius
    @radius.setter
    def radius(self, value):
        if value <= 0:
            raise TypeError(" 길이는 양이 숫자 ! ")
        self.__radius = value

#원의 둘레와 넓이
print("#데코레이터를 사용한 Getter, Setter")
circle = Circle(10)
print("원래 원의 반지름:", circle.radius)
circle.radius = 2
print("변경된 원의 반지름:", circle.radius)
print()

#강제로 예외 발생
print("#강제로 예외 발생")
circle.radius = -10

#error에 지정한 문구가 출력된 것 볼 수 있음
```

### 상속<br>
: 다른 누군가가 만들어 놓은 기본형태에 내가 원하는 것만 추가하거나 교체
- 다중상속 : 다른 누군가가 만들어 놓은 형태를 조립하여 내가 원하는 것을 만드는 것

- 부모 : 기반이 되는 것
- 자식 : 기반으로 생성한 것

```python
#상속의 활용

#부모 클래스 선언
class Parent:
    def __init__(self):
        self.value = "테스트"
        print("Parent클래스의 __init()__메소드 호출")
    def test(self):
        print("Parent 클래스의 test()메소드 호출")
    
#자식 클래스 선언
class Child(Parent): # 부모 상속
    def __init__(self):
        super().__init__() # 부모의 __init__()함수 호출
        print("Child클래스의 __init()__ 메소드 호출")

#자식 클래스의 인스턴스 생성, 부모 메소드 호출
child = Child()
child.test()
print(child.value)
```

### 예외 클래스 만들기
: CustomException 클래스 생성

```python
# 사용자정의 예외 클래스 생성
class CustomException(Exception):
    def __init__(self):
        super().__init__()

raise CustomException
```

```python
# 자식 클래스로써 부모의 함수 재정의(오버라이드)

class CustomException(Exception):
    def __init__(self):
        super().__init__()
        print("#### 내가 만든 오류 생성 ####")
    def __str__(self):
        return " 오류발생 "
    
raise CustomException
```

* 오버라이드/재정의 : 부모에 정의된 함수를 자식에서 다시 정의

* 기존에 있던 함수/변수 외에 새로 정의하는 것도 가능함

```python
#자식클래스로써 부모에 없는 새함수 정의
class CustomException(Exception):
    def __init__(self, message, value):
        super().__init__()
        self.message = message
        self.value = value

    def __str__(self):
        return self.message
    
    def print(self):
        print("### 오류정보 ###")
        print("메세지:", self.message)
        print("값:", self.value)

#예외발생시켜보기
try:
    raise CustomException("그냥해봄",273)
except CustomException as e:
    e.print()
```
