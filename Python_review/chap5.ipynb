```python
#손코딩_기본 매개변수와 키워드 매개변수를 활용해 범위의 정수를 더하는 함수

#함수선언
def sum_all(start=0, end=100, step=1):
    #변수선언
    output = 0
    #반복문을 돌려 숫자를 더합니다
    for i in range(start, end+1, step):
        output +=i
    #리턴
    return output

#함수호출
print("A.", sum_all(0,100,10))
print("B.", sum_all(end=100))
print("C.", sum_all(end=100,step=2))
```

```python
#5-2. 함수의 활용

#재귀함수
#팩토리얼 : n! = n * (n-1) * (n-2) * ... * 1
# - 반복문으로 팩토리얼 구하기
# - 재귀함수로 팩토리얼 구하기

#손코딩_ 반복문으로 팩토리얼 구하기
def factorial(n):
    #변수를 선언
    output = 1
    #반복문을 돌려 숫자더하기
    for i in range(1, n+1):
        output *= i
    #리턴
    return output

#함수호출
print("1!:", factorial(1))
print("2!:", factorial(2))
print("3!:", factorial(3))
print("4!:", factorial(4))
print("5!:", factorial(5))
```

```python
#재귀 : 자기자신을 호출하는 것
#n! = n * (n-1) * (n-2) * ... * 1
#factorial(n)=n*factorial(n-1) (n>=1일 때)
#factorial(0)=1

#f(4)=4*f(3)
#    =4*3*f(2)
#    =4*3*2*f(1)*f(0) -- f(0)=1
#    =4*3*2*1*1

#손코딩_재귀함수를 이용해 팩토리얼 구하기
#함수선언
def factorial(n):
    #n이 0이라면 1을 리턴
    if n ==0:
        return 1
    #n이 0이 아니라면 n*(n-1)!을 리턴
    else:
        return n * factorial(n-1)
    
#함수호출
print("1!:",factorial(1))
print("2!:",factorial(2))
print("3!:",factorial(3))
print("4!:",factorial(4))
print("5!:",factorial(5))
```

```python
#재귀함수의 문제
#재귀 함수는 상황에 따라서 같은 것을 기하급수적으로 많이 반복하는 문제 (+)
#해결방법 _ 메모화 기술

#재귀 함수로 구현한 피보나치 수열(1) _ 앞의 두 수를 더해서 다음 수를 구하는 방식
#함수선언
def fibonacci(n):
    if n == 1:
        return 1
    if n == 2:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
#함수 호출
print( "fibonacci(1):", fibonacci(1))
print( "fibonacci(2):", fibonacci(2))
print( "fibonacci(3):", fibonacci(3))
print( "fibonacci(4):", fibonacci(4))
print( "fibonacci(5):", fibonacci(5))
```

```python
#메모화

#같은 값을 구하는 연산을 반복 -> 같은 값을 한 번만 계산하도록 코드 수정

#손코딩_메모화
#메모변수 생성
dictionary = {
    1: 1,
    2: 1
}

#함수선언
def fibonacci(n):
    if n in dictionary:
        #메모가 되어 있으면 메모된 값 리턴
        return dictionary[n]
    else:
        #메모가 되어 있지 않으면 값을 구함
        output = fibonacci(n-1) + fibonacci(n-2)
        dictionary[n] = output
        return output

#함수 호출
print("fibonacci(10):", fibonacci(10))
print("fibonacci(20):", fibonacci(20))
print("fibonacci(30):", fibonacci(30))
print("fibonacci(40):", fibonacci(40))
print("fibonacci(50):", fibonacci(50))


#메모 : 딕셔너리를 사용해서 한 번 계산한 값 저장
#딕셔너리에 값이 메모되어 있으면 처리를 수행하지 않고 곧바로 메모된 값을 돌려주면서 코드 속도 빠르게
#메모화 사용하면 실행 속도 빨라짐
#재귀함수에 많이 사용되는 기술
```

```python
#filter()함수와 map()함수 - 함수를 매개변수로 사용하는 대표적인 표준 함수
# = 내장함수, 표준함수

#map() :리스트 요소를 함수에 넣고 리턴된 값으로 새로운 리스트 구성해주는 함수
#  map(함수, 리스트)

#filter() : 리스트의 요소를 함수에 넣고 리턴된 값이 True인 것으로 새로운 리스트를 구성하는 함수
#  filter(함수,리스트)

#손코딩_map(),filter()
#함수선언
def power(item):
    return item * item
def under_3(item):
    return item < 3

#변수선언
list_input_a = [1,2,3,4,5]

#map()함수사용
output_a = map(power,list_input_a) #함수 매개변수
print("# map() 함수의 실행 결과")
print("map(power, list_input_a):", output_a)
print("map(power, list_input_a):", list(output_a))
print()

#filter()함수 사용
output_b = filter(under_3, list_input_a) #함수 매개변수
print("# filter()함수 실행결과")
print("filter(under_3, list_input_a):",output_b)
print("filter(under_3, list_input_a):", list(output_b))
```

map()
- 첫번째 매개변수 > 함수 (값을 제곱해주는 power함수)
- 두번째 매개변수 > 리스트
-> power함수가 적용된 값을 반환

filter()
- 첫번째 매개변수 > ( item < 3을 판정하는 under_3()함수 )
- 두번째 매개변수 > 리스트
-> 리스트 내부 요소 중 item < 3 값이 True인 값만 반환

=> 제터레이터 : 두 결과로 나오는 <map object>, <filter object>
=> list()함수 적용하여 강제로 리스트 자료형으로 변환 후 출력함

```python
#손코딩_랜덤하게 1000명의 키와 몸무게를 만들기

#랜덤한 숫자를 만들기 위해 가져옴
import random

#간단한 한글 리스트를 만듬
hanguls = list("가나다라마바사아자차카타파하")

#파일을 쓰기 모드로 오픈
with open("info.txt", "w") as file:
    for i in range(1000):
        #랜덤한 값으로 변수 생성
        name = random.choice(hanguls) + random.choice(hanguls)
        weight = random.randrange(40,100)
        height = random.randrange(140, 200)
        #텍스트 쓰기
        file.write("{}, {}, {}\n".format(name, weight, height))

#info.txt파일에 생성된 것 확인 done !
```

```python
#반복문으로 파일 한 줄씩 읽기
with open("info.txt", "r") as file:
    for line in file:
        #변수선언
        (name, weight, height) = line.strip().split(", ")

        #데이터가 문제없는지 확인 : 문제가 있으면 지나감
        if (not name) or (not weight) or (not height):
            continue
        #결과계산
        bmi = int(weight) / ((int(height)) / 100 ** 2)
        result = ""
        if 25 <= bmi:
            result = "과체중"
        elif 18.5 <= bmi:
            result = "정상체중"
        else:
            result = "저체중"

        #출력
        print('\n'.join([
            "이름: {}",
            "몸무게: {}",
            "키: {}",
            "BMI: {}",
            "결과: {}"
        ]).format(name, weight, height, bmi, result))
        print()
```

제너레이터 : 파이썬만의 특수한 문법<br>
    - 이터레이터를 직접 만들 떄 사용하는 코드<br>
    - 함수 내부에 yield 키워드를 사용하면 해당 함수는 제너레이터 함수가 됨 -> 함수 호출해도 함수내부의 코드 실행 x (일반함수는 됬었자나 근데 얜 안된대)

```python
#손코딩_제너레이터 함수
#함수선언
def test():
    print("함수가 호출되었습니다.")
    yield " test"

#함수호출
print("A지점 통과")
test()

print("B 지점 통과")
test()
print(test())

#함수의 리턴값으로 "함수가 호출되었습니다."라는 문자열 출력 대신 <generator object test at 0x000002978BDA8E40> 출력
#   -> 제너레이터 함수는 제너레이터를 리턴
#   - <generator object test at 0x000002978BDA8E40> : 제너레이터 객체
```

```python
#제너레이터 객체와 next()함수
# 함수선언
def test():
    print("A 지점 통과")
    yield 1
    print("B지점 통과")
    yield 2
    print("C 지점 통과")
#함수호출
output = test()
#next() 함수 호출
print("D지점 통과")
a = next(output)
print(a)
print("E지점 통과")
b = next(output)
print(b)
print("F 지점 통과")
c = next(output)
print(c)
#한 번 더 실행
next(output)

#StipIteration 예외발생
#  -> 함수의 코드르 조금씩 실행할 때 사용 ( 메모리 효율성 )
```

리스트 함수의 key 키워드 매개변수<br>
    - 딕셔너리를 활용하여 가격이 가장 싼/비싼 것을 찾고 싶을 때 : 딕셔너리에 있는 키를 활용하면 됨!<br>
    - min()와 max()함수에 비교대상 key 키워드 매개변수 지정하여 최댓값, 최솟값 비교

```python
#손코딩_키워드 매개변수에 함수전달
#딕셔너리 내부에 있는 특성 속성을 활용하여 최솟값, 최댓값 비교
books = [{
    "제목": "혼공파",
    "가격": 18000
}, {"제목": "혼공머",
    "가격": 26000
}, {"제목": "혼공자",
    "가격": 24000
}]

def 가격추출함수(book):
    return book["가격"]

print("# 가장 저렴한 책")
print(min(books, key=가격추출함수))
print()

print("# 가장 비싼 책")
print(max(books, key=가격추출함수))
```

```python
#손코딩_ 콜백함수를 람다함수로 바꾸기
books = [{
    "제목": "혼공파",
    "가격": 18000
}, {"제목": "혼공머",
    "가격": 26000
}, {"제목": "혼공자",
    "가격": 24000
}]

print("# 가장 저렴한 책")
print(min(books, key=lambda book: book["가격"]))
print()

print("# 가장 비싼 책")
print(max(books, key=lambda book: book["가격"]))
```

```python
#객체 자료형 복사
def object_change1(d):
    d.append(4)

c=[1,2,3]

print(c)
object_change1(c)
print(c)

#메모리에 리스트 [1,2,3]만들고 -> c는 그 리스트 주소를 가리킴
#   c는 그 리스트 자체가 아니고 그 리스트가 저장된 주소(위치)를 가지고 있음
#object_change1(c)로 함수호출을 하면 c가 전달되는게 아니라
#   -> c가 가리키고 있는 리스트의 주소가 매개변수 d에 복사됨
# c → [1, 2, 3] ← d
```
