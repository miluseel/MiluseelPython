```python
#손코딩_기본 매개변수와 키워드 매개변수를 활용해 범위의 정수를 더하는 함수

#함수선언
def sum_all(start=0, end=100, step=1):
    #변수선언
    output = 0
    #반복문을 돌려 숫자를 더합니다
    for i in range(start, end+1, step):
        output +=i
    #리턴
    return output

#함수호출
print("A.", sum_all(0,100,10))
print("B.", sum_all(end=100))
print("C.", sum_all(end=100,step=2))
```

```python
#5-2. 함수의 활용

#재귀함수
#팩토리얼 : n! = n * (n-1) * (n-2) * ... * 1
# - 반복문으로 팩토리얼 구하기
# - 재귀함수로 팩토리얼 구하기

#손코딩_ 반복문으로 팩토리얼 구하기
def factorial(n):
    #변수를 선언
    output = 1
    #반복문을 돌려 숫자더하기
    for i in range(1, n+1):
        output *= i
    #리턴
    return output

#함수호출
print("1!:", factorial(1))
print("2!:", factorial(2))
print("3!:", factorial(3))
print("4!:", factorial(4))
print("5!:", factorial(5))
```

```python
#재귀 : 자기자신을 호출하는 것
#n! = n * (n-1) * (n-2) * ... * 1
#factorial(n)=n*factorial(n-1) (n>=1일 때)
#factorial(0)=1

#f(4)=4*f(3)
#    =4*3*f(2)
#    =4*3*2*f(1)*f(0) -- f(0)=1
#    =4*3*2*1*1

#손코딩_재귀함수를 이용해 팩토리얼 구하기
#함수선언
def factorial(n):
    #n이 0이라면 1을 리턴
    if n ==0:
        return 1
    #n이 0이 아니라면 n*(n-1)!을 리턴
    else:
        return n * factorial(n-1)
    
#함수호출
print("1!:",factorial(1))
print("2!:",factorial(2))
print("3!:",factorial(3))
print("4!:",factorial(4))
print("5!:",factorial(5))
```

```python
#재귀함수의 문제
#재귀 함수는 상황에 따라서 같은 것을 기하급수적으로 많이 반복하는 문제 (+)
#해결방법 _ 메모화 기술

#재귀 함수로 구현한 피보나치 수열(1) _ 앞의 두 수를 더해서 다음 수를 구하는 방식
#함수선언
def fibonacci(n):
    if n == 1:
        return 1
    if n == 2:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
#함수 호출
print( "fibonacci(1):", fibonacci(1))
print( "fibonacci(2):", fibonacci(2))
print( "fibonacci(3):", fibonacci(3))
print( "fibonacci(4):", fibonacci(4))
print( "fibonacci(5):", fibonacci(5))
```

```python
#메모화

#같은 값을 구하는 연산을 반복 -> 같은 값을 한 번만 계산하도록 코드 수정

#손코딩_메모화
#메모변수 생성
dictionary = {
    1: 1,
    2: 1
}

#함수선언
def fibonacci(n):
    if n in dictionary:
        #메모가 되어 있으면 메모된 값 리턴
        return dictionary[n]
    else:
        #메모가 되어 있지 않으면 값을 구함
        output = fibonacci(n-1) + fibonacci(n-2)
        dictionary[n] = output
        return output

#함수 호출
print("fibonacci(10):", fibonacci(10))
print("fibonacci(20):", fibonacci(20))
print("fibonacci(30):", fibonacci(30))
print("fibonacci(40):", fibonacci(40))
print("fibonacci(50):", fibonacci(50))


#메모 : 딕셔너리를 사용해서 한 번 계산한 값 저장
#딕셔너리에 값이 메모되어 있으면 처리를 수행하지 않고 곧바로 메모된 값을 돌려주면서 코드 속도 빠르게
#메모화 사용하면 실행 속도 빨라짐
#재귀함수에 많이 사용되는 기술
```

```python
#리스트 평탄화(2)
def flatten(data):
    output = []
    for item in data:
        if type(item) == list:
            output += flatten(item)
        else:
            output.append(item)
    return output

example = [[1,2,3],[4,[5,6]],7,[8,9]]
print("원본: ", example)
print("변환: ", flatten(example))
```

```python
#output.append(item)을 output += item으로 수정시 왜 안되나?
# -> append()함수와 += 연산자는 비슷해 보이지만 서로 다른기능을 함 

#append()함수는 매개변수 하나를 리스트에 마지막 요소로 추가
#_append(리스트_자료형)형태로 리스트를 전달하면 리스트 전체가 하나의 요소로서 추가됨

a=[1,2,3]
b=[4,5,6]
a.append(b)
print(a)

#+=연산자는 오른쪽에 있는 리스트의 요소를 하나하나 리스트에 추가
a = [1,2,3]
b = [4,5,6]
a += b
print(a)

# +=연산자와 같은 기능을 수행하는 함수 : extend()
# extend()를 다음과 같은 형태로 사용해 보면 +=연산자 코드와 같은 실행 결과를 냄
a = [1,2,3]
b = [4,5,6]
a.extend(b)
print(a)

#기억해야 할 사항
#1. 함수의 변수는 함수 호출마다 따로따로 만들어짐 (아웃풋 계속 이어지는 아님 !)
#2. 함수가 끝나면(리턴되면) 함수를 호출했던 위치로 돌아옴 (재귀적 !)
#output += flatten(items) => output += [5,6] => output += [4,5,6]
```

```python
#가독성 좋은 코드 작성 
# 1. 주석 '잘' 사용
# 2. 함수활용하여 코드 만들기 (def)

#코드유지보수
#3.14를 숫자로 입력한 상태
def get_circumference(radius):
    return 2 * 3.14 * radius
def get_circle_area(radius):
    return 3.14 * radius * radius

#3.14를 PI라는 변수로 설정한 상태 ( 가독성 높이기 )
#정의식의 3.14값을 변경해야 할 때 전체를 수정하지 않아도 됨
PI = 3.14

def get_circumference(radius):
    return 2 * PI * radius
def get_circle_area(radius):
    return PI * radius * radius

#함수를 사용하지 않은 경우 -> p태그로 감싸는 경우
#모든 코드를 수정하지 않아도 수정이 용이
#p 태그로 감싸는 함수
def p (content) :
    # 기존 코드 주석처리
    # return "<p>{}</p>".format(content)
    # 2022. 03.02 - 요청반영
    return "<p class= 'content-line'>{}</p>".format(content)

#출력
print(p("안녕하세요."))
print(p("간단한 HTML 태그를 만드는 예입니다."))
```
