### __name__ = "__main__"
* 엔트리 포인트 / 메인 : 프로그래밍 언어에서 프로그램의 진입점

    - __name__변수에 들어있는 값 = "__main__"


* 엔트리 포인트
: 프로그램이 시작되는 지점 - 파이썬이 이 파일부터 실행하겠다고 선택하는 곳

### 패키지
* 모듈이 모여서 구조를 이룬 것
* 모듈: import로 가져오는 모든 것 
* pip : python package index - 패키지 관리 시스템

```python
#패키지 내부 모듈 읽어오기
import test_package.module_a as a
import test_package.module_b as b 

print(a.variable_a)
print(b.variable_b)
```

* __init__.py파일 <br>
: 패키지를 읽을 때 어떤 처리를 수행하거나 패키지 내부 모듈 한번에 가져오고 싶을 때 
- 해당 폴더가 패키지 임을 알려주고, 패키지와 관련된 초기화 처리하는 파일
- __init__.py 에 __all__리스트 만듬 <br>
    -> 리스트에 지정한 모듈들이 from<패키지 이름> import * 할 때 전부 읽어들임

```python
# __init__.py
__all__ = ["module_a", "module_b"]

print("test_package")
```

### 텍스트데이터 / 바이너리 데이터
- 텍스트 데이터 : 사람이 읽을 수 있는 형태의 데이터 
- 바이너리 데이터 : 컴퓨터가 이해할 수 있는 형태의 데이터 
    - 이진숫자 : 컴퓨터가 내부적으로 처리하는 0, 1 숫자

### 인코딩 / 디코딩
- 인코딩 : 텍스트데이터 -> 글자 / 바이너리 데이터 -> 이미지 로 변환하는 방식
   - 텍스트데이터 : ASCII, UTF-16, EUC-KR, Shift-JIS 등
   - 바이너리 데이터 : JPEG, PNG, GIF 등
- 디코딩 : 인코딩된 데이터를 반대로 돌리는 방식 

```python
#이미지 읽어들이고 저장하기
from urllib import request

#구글 메인페이지 읽기
target = request.urlopen("https://www.hanbit.co.kr/images/common/logo_hanbit.png")
output = target.read()
print(output)

#write binary모드로
file = open("output.png", "wb")
file.write(output)
file.close()

#->forbidden... 제대로 출력될 시 b'로 감싸져 바이너리 데이터 임을 확인할 수 있음
```
