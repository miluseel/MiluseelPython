```python
#짝수와 홀수 구분하기

#if 조건문 기본형식
#if 불 값이 나오는 표현식:
#    불 값이 참일 때 실행할 문장

# 핵심은 if 문의 불 값이 나오는 조건문을 어떻게 만들 것인가 !

# 손코딩_ 끝자리로 짝수와 홀수 구분하기

#입력을 받습니다.
number = input("정수 입력>")

#마지막 자리 숫자를 추출
last_character = number[-1]

#숫자로 변환하기
last_number = int(last_character)

#짝수 확인
if last_number == 0\
    or last_number == 2 \
    or last_number == 4 \
    or last_number == 6 \
    or last_number == 8:
    print("짝수입니다")

#홀수 확인
if last_number == 1\
    or last_number == 3\
    or last_number == 5\
    or last_number == 7\
    or last_number == 9:
    print("홀수입니다")

#줄이 너무 길어질 떄는 \기호를 입력하고 줄바꿈해서 코드 입력
    
```

```python
#in연산자를 이용하여 문자열 내부에 찾고자 하는 문자열이 있는지 확인

#손코딩_in문자열 연산자를 활용하여 짝수와 홀수 구분
#문자열로 변환했으므로 last_character로 조건절 걸어야 함

#입력을 받습니다.
number=input("정수입력>")
last_character = number [-1]

#짝수조건
if last_character in "02468":
    print("짝수입니다")

#홀수조건
if last_character in "13579":
    print("홀수입니다.")
```

```python
#3-2. if~else와 elif구분

#else조건문의 활용
#if 조건문의 조건이 거짓일 떄 실행되는 부분

#if조건:
#   조건이 참일 때 실행할 문장
#else:
#   조건이 거짓일 떄 실행할 문장

#손코딩_if 조건문에 else구문을 추가해서 짝수와 홀수 구분하기

#입력을 받습니다.
number= input("정수 입력>")
number = int(number)

#조건문을 사용합니다
if number %2 ==0:
    #조건이 참일 때, 즉 짝수조건
    print("짝수입니다")
else:
    #조건이 거짓일 떄, 즉 홀수조건
    print("홀수입니다")

#조건이 오로지 두가지로만 구분될 때, if else구문을 사용하면 조건을 한번만 비교하므로 효율적
```

```python
#elif구문 _ 두가지로만 구분되지 않을 때 ( 계절, 요일 등 )

#if 조건A:
#    조건A가 참일 때 실행할 문장
#elif 조건B:
#   조건B가 참일 때 실행할 문장
#elif 조건c:
#   조건c가 참일 때 실행할 문장
#...
#else:
#   모든 조건이 거짓일 때 문장

#손코딩_계절구하기

#날짜/시간과 관련된 기능을 가져옵니다
import datetime

#현재 날짜/시간을 구하고
#쉽게 사용할 수 있게 월을 변수로 저장합니다
now = datetime.datetime.now()
month = now.month

#조건문으로 계절을 확인합니다
if 3<=month <=5:
    print("현재는 봄입니다")
elif 6<= month <= 8:
    print("현재는 여름입니다")
elif 9 <= month <= 11:
    print("현재는 가을입니다")
else:
    print("현재는 겨울입니다")
```

```python
#if 조건문을 효율적으로 사용하기
#조건문에서 이미 한번 비교한 것은 제외하고 다음으로 넘어감
#앞 단계 조건문에서 비교했떤 것을 다음 단계에서 한번 더 하고 있지 않는지 확인 !

#손코딩_유머를 조건문을 구현하기

#변수를 선언합니다
score = float(input("학점입력>"))

#조건문을 적용합니다
if score == 4.5:
    print("신")
elif 4.2 <= score :
    print("교수님의 사랑")
elif 3.5<= score:
    print("현 체제의 수호자")
elif 2.8 <= score :
    print("일반인")
elif 2.3 <= score :
    print("일탈을 꿈꾸는 소시민")
else:
    print("그냥 딴 일 찾아보자 우리")
```
