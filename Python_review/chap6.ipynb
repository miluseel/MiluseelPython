#기본예외처리<br>

예외처리 : 예외를 해결하는 모든 것<br>
    - 조건문사용<br>
    - try 구문 사용

```python
#예외상황확인

#예외가 발생할 수 있는 코드 
#숫자입력
number_input_a = int(input("정수 입력>"))

#출력
print("원의 반지름:", number_input_a)
print("원의 둘레:", 2 * 3.14 * number_input_a)
print("원의 넓이:", 3.14 * number_input_a * number_input_a)

#정수를 입력하지 않는다면, 오류발생
```

#조건문으로 예외처리<br> 
   -> 이전 상황에서 isdigit()사용하여 숫자로만 구성된 글자인지 확인 후 int()이용해 <br> 
      숫자로 변환할 수 없는 문자열 변환하면서 발생하는 예외 피할 수 있음

```python
#손코딩_조건문으로 예외 처리
user_input_a =input("정수 입력> ")

#사용자 입력이 숫자로만 구성되어 있을 때
if user_input_a.isdigit():
    #숫자로 변환
    number_input_a = int(user_input_a)
    #출력
    print("원의 반지름:", number_input_a)
    print("원의 둘레:", 2 * 3.14 *number_input_a)
    print("원의 넓이:", 3.14 * number_input_a * number_input_a)
else:
    print("정수를 입력하지 않았습니다.")

#프로그램 작성시 항상 예외적인 상황까지 모드 생각하는 습관 !
```

#try except구문<br> 
try:<br> 
    예외가 발생할 가능성이 있는 코드<br> 
except:<br> 
    예외가 발생했을 때 실행할 코드

```python
#손코딩_try except 구문
#try except구문으로 예외처리
try:
    #숫자로 변환
    number_input_a = int(input("정수 입력>")) #예외 발생 가능성 있는 곳
    #출력
    print("원의 반지름:", number_input_a)
    print("원의 둘레:", 2 * 3.14 *number_input_a)
    print("원의 넓이:", 3.14 * number_input_a * number_input_a)
except:
    print("무언가 잘못되었습니다.") #예외가 발생했을 때 실행할 구문
```

# try except 구문 + pass키워드 조합
- 예외 발생 코드 부분이 중요하지 않으면 프로그램 강제종료를 막기위한 목적으로 except구문에 아무것도 넣지 않고 try구문 사용함
    -> 아무것도 넣지 않으면 구문오류가 발생하므로 pass 키워드 넣어줌

try:
    예외가 발생할 가능성이 있는 코드
except:
    pass

```python
#숫자로 변환되는 것들만 리스트에 넣기
#변수선언
list_input_a = ["52","273","32","스파이","103"]

#반복적용
list_number = []
for item in list_input_a:
    #숫자로 변환해서 리스트에 추가
    try:
        float(item) #예외가 발생하면 알아서 다음으로 진행은 안되겠지?
        list_number.append(item) #예외없이 통과했으면 list_number 리스트에 넣그라
    except:
        pass

#출력
print("{}내부에 있는 숫자는".format(list_input_a))
print("{}입니다.".format(list_number))
```

#try except else구문 : 예외가 발생하지 않았을 떄 실행할 코드<br>
try:<br>
    예외가 발생할 가능성이 있는 코드<br>
except:<br>
    예외가 발생했을 때 실행할 코드<br>
else:<br>
    예외가 발생하지 않았을 때 실행할 코드

```python
#try except else 구문으로 예외처리
try:
    #숫자변환
    number_input_a = int(input("정수 입력>"))
except:
    print("정수를 입력하지 않았습니다.")
else:
    #출력
    print("원의 반지름:", number_input_a)
    print("원의 둘레:", 2 * 3.14 *number_input_a)
    print("원의 넓이:", 3.14 * number_input_a * number_input_a)

#try구문에 모두 넣고 처리해도 되긴 함
```

#finally구문 : 예외가 발생하든 발생하지 않든 무조건 실행할 때 사용하는 코드<br>
try:<br>
    예외가 발생할 가능성이 있는 코드<br>
except:<br>
    예외가 발생했을 때 실행할 코드<br>
else:<br>
    예외가 발생하지 않았을 때 실행할 코드<br>
finally:<br>
    무조건 실행할 코드

```python
#finally구문
#try except 구문으로 예외 처리
try:
    #숫자변환
    number_input_a = int(input("정수 입력>"))
   #출력
    print("원의 반지름:", number_input_a)
    print("원의 둘레:", 2 * 3.14 *number_input_a)
    print("원의 넓이:", 3.14 * number_input_a * number_input_a)
except:
    print("정수를 입력하지 않았습니다.")
else:
    print("예외가 발생하지 않았습니다.")
finally:
    print("일단 프로그램이 어떻게든 끝났습니다.") # 두가지 경우 모드 실행됨
```

#try, except, finally 구문조합 규칙<br>
    - try 구문은 단독사용 안됨, 반드시 except구문/finally구문이랑 함께 사용<br>
    - else 구문은 반드시 except 구문 뒤에 사용<br>
        - try + except 구문 조합<br>
        - try + except + else 구문 조합<br>
        - try + except + finally 구문 조합<br>
        - try + except + else + finally 구문 조합<br>
        - try + finally 구문 조합<br>
        이 외에는 구문오류

```python
#finally 구문 이용해 파일 닫기
# try excpet구문
try:
    #파일열기
    file = open("info.txt", "w")
    #여러가지 처리 수행
    예외.발생해라ㅋㅋㅋ()
except:
    print("오류가 발생했습니다.")
finally:
    #파일을 닫습니다.
    file.close()
print("#파일이 제대로 닫혔니? 함봐라 얘야")
print("file.closed:", file.closed)

#try except구문 끝난 후 파일 닫으면 문제 없음
#그러므로 무조건 finally써야 하는 것은 아님
```

```python
#finally 구문은 반복문 / 함수 내부에 있을 때 위력발휘 !

#try구문 내부에서 return키워드 사용
#test()함수 선언
def test():
    print("test() 함수의 첫 줄입니다.")
    try:
        print("try 구문이 실행되었습니다.")
        return
        print("try구문의 return 키워드 뒤 입니다.")
    except:
        print("except구문이 실행되었다")
    else:
        print("else 구문 실행됨")
    finally:                        #finally구문은 무조건 실행되는거 아시져?? return으로 중간탈출해도 무조건입니다요
        print("finally구문 실행")
    print("test()함수의 마지막줄!!")

#test()함수호출
test()
```

```python
#finally키워드 활용
#함수선언
def write_text_file(filename, text):
    #try except구문 사용
    try:
        #파일열기
        file = open(filename, "w")
        #여러가지 처리 수행
        return
        #파일에 텍스트 입력
        file.write(text)
    except:
        print("오류발생")
    finally:
        #파일닫기
        file.close()
#함수호출
write_text_file("test.txt", "안녕하세요!")

#return으로 빠져나갈 때마다 close()하면 복잡해지는데, finally구문에서 close()호출하면 깰꼼
```

#반복문과 함께 사용할 때 - break로 빠져나가도 마찬가지로 finally구문은 무조건 실행

```python
#손코딩_반복문과 함께 사용하는 경우
print("프로그램이 시작되었습니다.")

while True:
    try:
        print("try 구문 실행")
        break
        print("try 구문의 break키워드 뒤다~ 이말이야")
    except:
        print("except 구문 실행")
    finally:
        print(("여긴 finally 구문이 실행됬다~ 이말이야"))
    print("while반복문의 마지막 줄 도달!")
print("프로그램 종료")

#break로 구문빠져나가도 finally는 실행 됬다는 것을 알 수 있음 
```
