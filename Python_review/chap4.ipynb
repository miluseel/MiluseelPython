```python
#리스트 요소 제거 (2) _ 인덱스로 제거 / 값으로 제거

# 인덱스로 제거
# 1. del키워드 
# del 라스트명[인덱스]
# 하나만 제거

# 2. pop()
# 리스트명.pop(인덱스)
# 퐁당퐁당으로 제거

list_a = [0,1,2,3,4,5]
print("# 리스트의 요소 하나 제거하기")

# 제거방법[1] - del
del list_a[1]
print("del list_a[1]:",list_a)

# 제거방법[2] - pop()
list_a.pop(2)
print("pop(2):",list_a)
```

```python
#get() : 딕셔너리의 키로 값을 추출하는 기능
#딕셔너리[키]를 입력할 떄랑 같은 기능 수행 / 존재하지 않는 키에 접근 할 경우 keyError가 아닌 None출력

#딕셔너리 선언
dictionary = {
    "name": "7D 건조 망고",
    "type": "당절임",
    "ingredient": ["망고","설탕","메타종아황산나트륨","치자황색소"],
    "orgin":"필리핀"
}

#존재하지 않는 키에 접근
value = dictionary.get("존재하지 않는 키")
print("값:",value)

#None 확인 방법
if value == None:
    print("존재하지 않는 키에 접근했습니다.")
```

```python
#손코딩_중첩반복문으로 피라미드 만들기

output = ""

for i in range(1,10):
    for j in range(0,i):
        output +="*" #내부 반복문: 별생성
    output += "\n" #외부 반복문: 줄생성

print(output)
```

```python
#손코딩_반복문으로 피라미드 만들기(2)

output=""

for i in range(1,15): 
    for j in range(14, i, -1):  #공백 생성 _전체 높이 (4)-j
        output +=' '
    for k in range(0,2*i-1):#별 생성 -_2k-1개
        output +='*'
    output +='\n' #/n _ 줄생성

print(output)
```

```python
#문자열 * 연산자로 피라미드 만들기 ( 더 쉬운 방법 )

output = ""

for i in range(1,10):
    output += ("*"*i)
    output += "\n"

print(output)
```

```python
#while 반복문 : 상태를 기반으로 반복
#리스트의 remove()는 리스트 내부에서 해당하는 값을 하나만 제거할 수 있었음
#while 반복문은 여러개 제거 가능

#손코딩_ 해당하는 값 모두 제거
#변수를 선언
list_test = [1,2,1,2]
value = 2

#list_test 내부에 value 있다면 반복
while value in list_test:
    list_test.remove(value)

#출력
print(list_test)
```

```python
#while 반복문 : 시간 기반으로 반복
#유닉스 타임: 세계표준시(UTC) 기준으로 몇 초가 지났는지를 정수로 나타낸 것
#시간과 관련된 기능 가져오기 -> 유닉스타임 구함

#손코딩_5초동안 반복하기

#시간과 관련된 가능 가져옴
import time

#변수선언
number = 0

#5초동안 반복
target_tick = time.time() +5
while time.time() < target_tick:
    number +=1

#출력
print("5초동안 {}번 반복했습니다.".format(number))

#컴퓨터 성능과 상황에 따라 반복횟수 계속 달라짐
#통신할 때 자주 사용하는 코드
```

```python
#while 반복문 : break키워드/ continue키워드
#break :반복문을 벗어날 때 사용하는 키워드 _ 무한반복을 만들고, 내부반복을 벗어날 때 사용

#손코딩_break 키워드
#변수를 선언
i = 0

#무한반복
while True:
    #몇번째 반복인지 출력
    print("{}번째 반복문입니다.".format(i))
    i = i + 1
    #반복을 종료합니다.
    input_text = input(">종료하시겠습니까?(y/n): ")
    if input_text in ["y","Y"]:
        print("반복을 종료합니다.")
        break
```

```python
#continue키워드 : 현재 반복을 생략하고 다음 반복으로 넘어갈 때 사용하는 키워드

#손코딩_continue키워드
#변수를 선언
numbers = [5,15,6,20,7,25]

#반복을 돌립니다
for number in numbers:
    #number가 10보다 작으면 다음 반복으로 넘어갑니다.
    if number < 10:
        continue
    #출력
    print(number)
```

```python
#continue키워드 사용하지 않은 경우
numbers = [5,15,6,20,7,25]
#반복돌림
for number in numbers:
    #반복대상 한정
    if number >= 10:
        print(number)
```

```python
#enumerate()함수와 반복문 조합
#리스트의 요소를 반복할 때 현재 인덱스가 몇 번째인지 확인

#손코딩_enumerate()함수와 리스트

#변수선언
example_list = ["요소A","요소B","요소C"]

#그냥 출력
print("#단순 출력")
print(example_list)
print()

#enumerate()함수 적용 후 출력
print("#enumerate()함수 적용 출력")
print(enumerate(example_list))
print()

#list()함수로 강제 변환 후 출력
print("#list()함수로 강제 변환 출력")
print(list(enumerate(example_list)))
print()

#for반복문과 enumerate()함수 조합해서 사용
print("# 반복문과 조합하기")
for i, value in enumerate(example_list):
    print("{}번째 요소는 {}입니다.".format(i, value))
```

```python
#리스트 내포
#리스트 이름 = [표현식 for 반복자 in 반복할 수 있는 것]
#리스트 이름 = [표현식 for 반복자 in 반복할 수 있는 것 if 조건문]

#손코딩_조건을 활용한 리스트 내포

#리스트 선언
array =["사과","자두","초콜릿","바나나",'체리']
output = [fruit for fruit in array if fruit != "초콜릿"] # != 초콜렛 초콜렛이 아닌 리스트의 모든 것을 출력
          
#출력
print(output)
```

```python
#손코딩_여러 줄 문자열과 if구문을 조합했을 때의 문제 해결 (1)

#변수선언
number = int(input("정수 입력>"))

#if 조건문으로 홀수 짝수 구분
if number %2 ==0:
    print((
        "입력한 문자열은 {}입니다.\n"
        "{}는 짝수입니다."
    ).format(number, number))
else:
    print((
        "입력한 문자열은 {}입니다.\n" 
        "{}는 홀수입니다."
    ).format(number,number))
```

```python
#손코딩_여러 줄 문자열과 if구문을 조합했을 때의 문제 해결 (2)

if number %2 == 0:
    print("\n".join([
        "입력한 문자열은 {}입니다.",
        "{}는 짝수입니다."
    ]).format(number, number)) #문자열 사이에 \n으로 줄바꿈이 join됨
else:
    print("\n".join([
        "입력한 문자열은 {}입니다.",
        "{}는 홀수입니다."
    ]).format(number,number))
```

```python
#이터레이터
#for 반복자 in 반복할 수 있는 것
#이터러블 : 반복할 수 있는 것 _내부에 있는 요소들을 차례차례 꺼낼 수 있는 객체 (리스트, 딕셔너리, 문자열, 튜플 등 )
#이터레이터 : 이터러블 중에서 next()함수를 적용해 하나하나 꺼낼 수 있는 요소

#손코딩_ reversed()와 이터레이터

#변수선언
numbers = [1,2,3,4,5,6]
r_num = reversed(numbers)

#reversed_numbers출력
print("reversed_numbers:",r_num)
print(next(r_num))
print(next(r_num))
print(next(r_num))
print(next(r_num))
print(next(r_num))
print(next(r_num))
```
