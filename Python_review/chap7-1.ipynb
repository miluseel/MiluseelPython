### random 모듈
- 랜덤한 값을 생성할 때 사용하는 모듈
- from , as 구문과도 조합 가능
import random
- . 찍으면 모듈이 가진 기능 확인 가능

```python
#손코딩_random 모듈
import random
print("# random 모듈")

#random(): 0.0 <= x < 1.0 사이의 float리턴
print("- random():", random.random())

#uniform(min, max):지정한 범위 사이의 float리턴
print("- uniform(10,20):", random.uniform(10,20))

#randrange(): 지정한 범위의 int리턴
# - randrange(max) :0부터 max 사이의 값 리턴
# - randrange(min,max) : min부터 max사이의 값 리턴
print("- randrange(10):",random.randrange(10))

#choice(list):리스트 내부에 있는 요소 랜덤하게 선택
print("- choice)[1,2,3,4,5]):",random.choice([1,2,3,4,5]))

#shuffle(list):리스트 요소들을 랜덤하게 섞기
print("- shuffle([1,2,3,4,5]):", random.shuffle([1,2,3,4,5]))

#sample([list, k=<숫자>]): 리스트 요소 중에 k개 뽑기
print("- sample([1,2,3,4,5],k=2):"), random.sample([1,2,3,4,5], k=2) 
```

### datetime모듈
: date, time과 관련된 모듈
- 날짜 형식을 만들 때 자주 사용되는 코드들로 구성

```python
#손코딩_datetime모듈

#모듈을 읽어들임
import datetime

#현재 시각을 구하고 출력
print("현재 시각 출력")
now = datetime.datetime.now()
print(now.year,"년")
print(now.month, "월")
print(now.day, "일")
print(now.hour,"시")
print(now.minute,"분")
print(now.second, "초")
print()

#시간출력
print("#시간을 포맷에 맞춰 출력")
output_a = now.strftime("%Y.%m.%d %H:%M:%S")
output_b = "{}년 {}월 {}일 {}시 {}분 {}초".format(now.year,\
    now.month,\
    now.day,\
    now.hour,\
    now.minute,\
    now.second)
output_c=now.strftime("%Y{} %m{} %d{} %H{} %M{} %S{}").format(*"년월일시분초") # 문자열, 리스트 등 앞에 * 붙이면 요소 하나하나가 매개변수로 저장
print(output_a)
print(output_b)
print(output_c)
print()
```

* 'strftime()함수' 사용시 시간을 형식에 맞춰 출력가능<br>
* 한국어 등 문자는 매개변수로 넣을 수 없음 -> output_b, output_c같은 형식으로 보완<br>

<보완방식><br>
output_b = "{}년 {}월 {}일 {}시 {}분 {}초".format(now.year,\
    now.month,\
    now.day,\
    now.hour,\
    now.minute,\
    now.second)
output_c=now.strftime("%Y{} %m{} %d{} %H{} %M{} %S{}").format(*"년월일시분초") <br># 문자열, 리스트 등 앞에 * 붙이면 요소 하나하나가 매개변수로 저장

```python
#손코딩_시간처리

#모듈읽기
import datetime
now = datetime.datetime.now()

#특정 시간 이후 시간 구하기
print("#datetime.timedelta로 시간 더하기")
after = now + datetime.timedelta(\
    weeks=1,\
    days=1,\
    hours=1,\
    minutes=1,\
    seconds=1)
print(after.strftime("%Y{} %m{} %d{} %H{} %M{} %S{}").format(*"년월일시분초"))
print()

#특정시간 요소 교체
print("#now.replace()로 1년 더하기")
output = now.replace(year=(now.year + 1))
print(output.strftime("%Y{} %m{} %d{} %H{} %M{} %S{}").format(*"년월일시분초"))

```

* timedelta()함수 사용하면 특정한 시간의 이전 또는 이후 구할 수 있음<br>
* 1년 후 , 2년 후 등은 replace()함수 사용<br>

### operator.itemgetter()함수

```python
#더 알아보기_operator.itemgetter()
books = [{
    "제목": "혼공파",
    "가격": 18000
}, {
    "제목": "혼공머 + 딥러닝",
    "가격": 26000
}, {
    "제목": "혼공자",
    "가격": 24000
}] 

def 가격추출함수(book):
    return book["가격"]

min(books, key=가격추출함수) #콜백
min(books, key=lambda book: book["가격"]) #람다
```

operator모듈의 itemgetter()함수 : 특정 요소를 추출하는 함수를 만드는 함수
    - 코드를 읽으면서 바로 이해할 수 있음
    - 람다함수를 사용한 것보다 코드 읽는 것이 쉬움

```python
#key 매개변수를  itemgetter()함수로 변경
from operator import itemgetter

books = [{
    "제목": "혼공파",
    "가격": 18000
}, {
    "제목": "혼공머 + 딥러닝",
    "가격": 26000
}, {
    "제목": "혼공자",
    "가격": 24000
}] 

print("#가장 저렴한 책")
print(min(books, key=itemgetter("가격")))
print()

print("# 가장 비싼 책")
print(max(books, key=itemgetter("가격")))
```

```python
#확인문제3

#현재 디렉터리를 읽어 들이고 파일인지 디렉터리인지 구분
#모듈읽기
import os

#현재 폴더의 파일/폴더 출력
output = os.listdir(".")
print("os.listdir():", output)
print()

#현재 폴더의 파일/폴더 구분
print("#폴더와 파일 구분")
for path in output:
    if os.path.isdir(path):
        print("폴더:",path)
    else:
        print("파일:", path)
```

```python
#모듈읽어오기
import os

#폴더를 읽어 들이는 함수
def read_folder(path):
    #폴더요소 읽어들이기
    output = os.listdir(path)
    #폴더요소 구분
    for item in output:
        if os.path.isdir(item):
            #폴더라면 계속 읽기
            read_folder(item)
        else:
            #파일이면 출력
            print("파일:",item)

# 현재 폴더의 파일/폴더 출력
read_folder(".")
```
