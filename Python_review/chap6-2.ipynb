    예외 발생 가능성이 있는 구문<br>

###예외 객체 : Exception클래스


```python
#손코딩_예외객체
#try except구문으로 예외 처리
try:
    #숫자변환
    number_input_a = int(input("정수입력>"))
    #출력
    print("원의 반지름:" , number_input_a)
    print("원의 둘레:, 2 * 3.14 * number_input_a)
    print("원의 넓이:, 3.14 * number_input_a * number_input_a)
except Exception as exception:
    #예외 객체 출력
    print("type(exception):", type(exception))
    print("exception:", exception)


#예외를 강제로 발생시켰을 때 , 객체자료형은 Value Error
```

##예외구분<br>
예외 객체를 사용 - except구문을 if조건문처럼 사용해서 예외구분 가능

```python
#손코딩_여러가지 예외가 발생할 수 있는 상황

#변수선언
list_number = [ 52, 273, 32, 72, 100]

#try except 구문으로 예외처리
try:
    #숫자입력
    number_input = int(input("정수 입력>"))
    #리스트의 요소 출력
    print("{}번째 요소: {}".format(number_input, list_number[number_input]))
except Exception as exception:
    #예외 객체 출력
    print("type(eception):", type(exception))
    print("exception:", exception)
```

- 정수입력 : 결과 잘 반환
- 정수변환 불가값 : Value Error
- 리스트의 길이 넘는 인덱스 입력 : Index Error

##예외구분하기<br>
except구문 뒤에 예외의 종류 입력하여 예외구분<br>

try:<br>
    예외 발생 가능성 있는 구문<br>
except 예외의 종류A:<br>
    예외A가 발생했을 때 실행할 구문<br>
except 예외의 종류B:<br>
    예외가 발생했을 때 실행할 구문<br>
excpet 예외의 종류C:<br>
    예외가 발생했을 때 실행할 구문

```python
#손코딩_예외 구분하기
#변수선언
list_number = [52, 273, 32, 72, 100]

#try except구문으로 예외 처리
try:
    #숫자입력
    number_input = int(input("정수입력>"))
    #리스트 요소 출력
    print("{}번째 요소: {}".format(number_input, list_number[number_input]))
except ValueError:
    # ValueError발생
    print("정수를 입력하세요")
except IndexError:
    # IndexError발생
    print("리스트의 인덱스를 벗어남")
```

###예외 구분 구문과 예외 객체<br>
except 구문 뒤에 예외 객체 붙여 활용 : as 키워드

```python
#손코딩_예외 구분 구문과 예외 객체
#예외 객체 함께 출력

try:
    #숫자입력
    number_input = int(input("정수입력>"))
    #리스트 요소 출력
    print("{}번째 요소: {}".format(number_input, list_number[number_input]))
except ValueError as exception:
    # ValueError발생
    print("정수를 입력하세요")
    print("exception:", exception)
except IndexError as exception:
    # IndexError발생
    print("리스트의 인덱스를 벗어남")
    print("exception:",exception)
```

###모든 예외 잡기<br>
예외 발생으로 프로그램이 종료될 수 있음 -> else처럼 모든 예외를 포함하는 exception넣어 프로그램 죽지 않게

```python
#손코딩_모든 예외 처리 했지만 못 잡는 경우

try:
    #숫자입력
    number_input = int(input("정수입력>"))
    #리스트 요소 출력
    print("{}번째 요소: {}".format(number_input, list_number[number_input]))
    예외.발생() 
except ValueError as exception:
    # ValueError발생
    print("정수를 입력하세요")
    print("exception:", exception)
except IndexError as exception:
    # IndexError발생
    print("리스트의 인덱스를 벗어남")
    print("exception:",exception)
```

```python
#손코딩_ 모든 예외잡기

try:
    #숫자입력
    number_input = int(input("정수입력>"))
    #리스트 요소 출력
    print("{}번째 요소: {}".format(number_input, list_number[number_input]))
    예외.발생() 
except ValueError as exception:
    # ValueError발생
    print("정수를 입력하세요")
    print(type(exception), exception)
except IndexError as exception:
    # IndexError발생
    print("리스트의 인덱스를 벗어남")
    print(type(exception),exception)
except Exception as exception:
    # 이외의 예외 발생
    print("미리 파악하지 못한 예외 발생")
    print(type(exception),exception)


#type(exception)으로 바꾼 이유 : 오류의 타입을 출력하게 위해서 (프로그램이 강제종료되지 않는 상태에서 )
```
